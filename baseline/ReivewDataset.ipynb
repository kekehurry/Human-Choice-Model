```python
from models.data import get_dataset, get_data

train_file = 'data/train/Eat.csv'
test_file = 'data/test/Eat.csv'
train_dataset, test_dataset, mapping = get_dataset(
    train_file=train_file, test_file=test_file)

X_train, y_train = get_data(train_dataset, mapping, y_prop='target_mode')
```

```python
import pandas as pd
import numpy as np
data = pd.merge(X_train, y_train, on='person_id')
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
matrix = data.corr()


sns.heatmap(matrix, annot=True)
plt.show()
```
